# define core functions
Just some useful scripts that will make my machine learning project cleaner. You don't really have to try to understand it, but don't worry I won't cheat either.

```python
def create_dataset(_filenames, _label) :
    _master = pd.DataFrame()
    
    for _file in _filenames:
        _df = pd.read_csv(_file)
        _master = _master.append(_df)
    
    _master[_label[0]] = _label[1]
    return _master

def run_algorithm(_feature, _label, _classifier, **kwargs) :
    # construct kfold
    _kf    = KFold(len(_label), n_folds=10, shuffle=True)
    
    # create a list of predictions for output
    _yPred = _label.copy()
    
    # iterate through folds, assign training features/labels, then fit to the model
    for _idxTrain, _idxTest in _kf:
        _featureTrain, _featureTest = _feature[_idxTrain], _feature[_idxTest]
        _labelTrain   = _label[_idxTrain]
        
        _clf = _classifier(**kwargs)
        _clf.fit(_featureTrain, _labelTrain)
        _yPred[_idxTest] = _clf.predict(_featureTest)
        
    return _yPred

def construct_model(_X, _y, _classes, _name, kwargs=None) :
    _cls   = _classes if not isinstance(_name, basestring) else [_classes]
    kwargs = {} if kwargs is None else kwargs
    
    for _i, _class in enumerate(_cls):
        #get current start time
        _start = time.time()
        _pred  = run_algorithm(_X, _y, _class, **kwargs)
        
        _acc = np.mean(_y == _pred)
        _fs  = precision_recall_fscore_support(_y, _pred, average="binary")
        
        _time  = time.time() - _start
        _title = _name[_i] if not isinstance(_name, basestring) else _name
        
        print _title
        print "=" * len(_title)
        print "accuracy: %.3f" % _acc
        print "recall: %.3f" % _fs[0]
        print "precision: %.3f" % _fs[1]
        print "fscore: %.3f" % _fs[2]
        print "run time in seconds: %.3f" % _time
        print " "

def optimize_algorithm(_X, _y, _clf, _param) :
    for _try in range(_param['min'], _param['max']):
        _start = time.time()
        kwargs = {
            _param['kw'] : _try
        }
        _acc = np.mean(_y == run_algorithm(_X, _y, _clf, **kwargs))
        
        print "accuracy: %.3f" % _acc
        
# report model with best hyperparameter optimization 
def report(_gridScores, nTop=1):
    _topScores = sorted(_gridScores, key=itemgetter(1), reverse=True)[:nTop]
    for _i, _score in enumerate(_topScores):
        print("Model with rank #{0}".format(_i + 1))
        print("Mean validation score: {0:.3f} (std: {1:.3f})".format(_score.mean_validation_score, np.std(_score.cv_validation_scores)))
        print("Parameters: {0}".format(_score.parameters))
        print("")
```

# feature engineering
Let's talk some science. Feature engineering is about creating new features based on domain knowledge that would fit the machine learning model better, thus easier to classify and get better results. 

In our case, we know that a user's status of activity (active or inactive) heavily correlates with number of logins to the application. However due to the nature of this app, number of infographics create would better represent a user's activity compare to logins. A user who logs in to the app often but does nothing could mean he/she is away or did not achieve what the app should provide.

A new feature based on user's activity is created by looping each user and retrieve all infographics created by that user. Time difference is then calculated by taking the date create for current infographics and the previous infographic, and then the total time difference is average by hours. The average time difference is labeled as `hr_diff`.

Example, a user who created 5 infographics will have:

`time_difference = SUM(date_create_5th - date_create_4th, date_create_4th - date_create_3rd, ....) / 5`

The chart below shows the relationship between time difference of creating infographics and number of infographics created, segmented by active and inactive users. I've included a chart to show the relationship between login counts and number of infographics for comparison.

```python
import matplotlib.pyplot as plt
import seaborn

plt.figure(figsize=(20, 8))
plt.subplot(2, 2, 1)
plt.scatter(_active.hr_diff, _active.igs, c='#1C3BEB', label='active', alpha=0.5)
plt.scatter(_dormant.hr_diff, _dormant.igs, c='#ED284C', label='dormant', alpha=0.5)
plt.title("relationship between time difference and number of infographics")
plt.xlabel("time difference")
plt.ylabel("number of infographics")
plt.ylim(ymin= 5, ymax=100)
plt.xlim(xmin=-5)

plt.subplot(2, 2, 2)
plt.scatter(_active.hr_diff, _active.igs, c='#1C3BEB', label='active', alpha=0.5)
plt.title("relationship between time difference and number of infographics (active users only)")
plt.xlabel("time difference")
plt.ylabel("number of infographics")
plt.ylim(ymin= 5, ymax=100)
plt.xlim(xmin=-5)

plt.figure(figsize=(20, 8))
plt.subplot(2, 2, 3)
plt.scatter(_active.sign_in_count, _active.igs, c='#1C3BEB', label='active', alpha=0.5)
plt.scatter(_dormant.sign_in_count, _dormant.igs, c='#ED284C', label='dormant', alpha=0.5)
plt.title("relationship between login count and number of infographics")
plt.xlabel("login count")
plt.ylabel("number of infographics")
plt.ylim(ymin= 5, ymax=100)
plt.xlim(xmin=-5, xmax=200)

plt.subplot(2, 2, 4)
plt.scatter(_active.sign_in_count, _active.igs, c='#1C3BEB', label='active', alpha=0.5)
plt.title("relationship between login count and number of infographics (active users only)")
plt.xlabel("login count")
plt.ylabel("number of infographics")
plt.ylim(ymin= 5, ymax=100)
plt.xlim(xmin=-5, xmax=200)

plt.show()
```

#feature scaling with StandardScaler

Before we create our machine learning model, it is best to do feature scaling to normalize the range of values in the features. Datasets without standardization may cause the some errors in the algorithm calculation due to different weightage values for different features, thus you won't get an accurate result.

```python
# convert the labels to 0 and 1
_y = np.where(_ds.active == "True", 1, 0)

# standardization
from sklearn.preprocessing import StandardScaler

_X = _fs.as_matrix().astype(float)
_X = StandardScaler().fit_transform(_X)

print "feature space holds %d observations and %d features" % _X.shape
print "unique target labels:", np.unique(_y)
```

# applying supervised learning algorithms
We will test our dataset with 5 supervised learning algorithms: Naive Bayes, SVM, Decision Trees, Random Forest, K-nearest-neighbors. The idea is to run the same dataset on all 5 algorithms and see which algorithm produce the best result. The evaluation metrics will be the accuracy, recall, and precision.

## precision and recall
Shall we based our result on higher precision, or higher recall? The formula for precision and recall are as below.

`recall = true positive / (true positive + false negatives)`

`precision = true positive / (true positive + false positives)`

Assuming if we deploying this machine learning model, it should label a user whether active or inactive after certain period of time. Users who are labeled as active are potential 'leads' which agents can approach to upsell services. My rationale is it would be better label as user as 'active' even if user is actually inactive than fail to label an active user as 'active', thus missing the opportunity for upselling.

Therefore in this project, precision is better than recall.

```python
# set the ML algorithms and names into lists.
_class = [NB, SVC, DT, RF, KNN]
_names = ["Gaussian Naive Bayes", "Support Vector Machines", "Decision Tree", "Random Forest", "K-nearest-neighbors"]

construct_model(_X, _y, _class, _names)
```

RandomSearchCV would take alot lesser time to get the best parameter settings, however it samples random sets of parameters and may not get the actual best settings. Obviously, it's not feasible to test every parameter settings via GridSearchCV as well.
